```python
import wikipediaapi
from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
import psycopg2
import pandas as pd

# This code returns all the text inside each linked page on the main category page
wiki_wiki = wikipediaapi.Wikipedia('en')

cat_women_editors = wiki_wiki.page("Category:Women editors")
dict_keys_cat_women_editors = cat_women_editors.categorymembers.keys()

sql_queries = []

for woman in dict_keys_cat_women_editors:
    sql_queries.append(("INSERT INTO profiles_api (name, bio) VALUES (%s, %s)", woman, wiki_wiki.page(woman).text))
```
